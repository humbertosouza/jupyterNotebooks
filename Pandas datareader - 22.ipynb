```python
# Python 3 initialization
import pandas_datareader as pdweb

import datetime
```

```python
prices = pdweb.get_data_yahoo(['CVX','XOM','BP'], start=datetime.datetime(2015,1,1), 
                              end=datetime.datetime(2018,1,1))['Adj Close']
prices.head()
```

```python
volume= pdweb.get_data_yahoo(['CVX','XOM','BP'],start=datetime.datetime(2015,1,1), 
                             end=datetime.datetime(2018,1,1))['Volume']
```

```python
volume.head()
```

```python
rets = prices.pct_change()
```

```python
# correlation of the stocks
corr = rets.corr
```
